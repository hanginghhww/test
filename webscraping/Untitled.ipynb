```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
import re
%matplotlib inline

plt.style.use('ggplot')
```

```python
eye_mu=pd.read_csv('eye.csv',names=['brandname','productname','lowprice','highprice','rate'])
face_mu=pd.read_csv('face.csv',names=['brandname','productname','lowprice','highprice','rate'])
cheek_mu=pd.read_csv('cheek.csv',names=['brandname','productname','lowprice','highprice','rate'])
lip_mu=pd.read_csv('lip.csv',names=['brandname','productname','lowprice','highprice','rate'])
mo_sc=pd.read_csv('moisturizer.csv',names=['brandname','productname','lowprice','highprice','rate'])
cleanser_sc=pd.read_csv('cleanser.csv',names=['brandname','productname','lowprice','highprice','rate'])
treatment_sc=pd.read_csv('treatment.csv',names=['brandname','productname','lowprice','highprice','rate'])
tool_sc=pd.read_csv('tools.csv',names=['brandname','productname','lowprice','highprice','rate'])
mask_sc=pd.read_csv('mask.csv',names=['brandname','productname','lowprice','highprice','rate'])
eye_sc=pd.read_csv('eye_sc.csv',names=['brandname','productname','lowprice','highprice','rate'])

# s=[eye_mu,face_mu,cheek_mu,lip_mu,mo_sc,cleanser_sc,treatment_sc,tool_sc,mask_sc,eye_sc]
eye_mu['subcat']='eye'
eye_mu['cat']='makeup'
face_mu['subcat']='face'
face_mu['cat']='makeup'
cheek_mu['subcat']='cheek'
cheek_mu['cat']='makeup'
lip_mu['subcat']='lip'
lip_mu['cat']='makeup'
mo_sc['subcat']='moisturizer'
mo_sc['cat']='skincare'
cleanser_sc ['subcat']='cleanser'
cleanser_sc['cat']='skincare'
treatment_sc['subcat']='treatment'
treatment_sc['cat']='skincare'
tool_sc['subcat']='tool'
tool_sc['cat']='skincare'
mask_sc['subcat']='mask'
mask_sc['cat']='skincare'
eye_sc['subcat']='eye_sc'
eye_sc['cat']='skincare'
frames = [eye_mu,face_mu,cheek_mu,lip_mu,mo_sc,cleanser_sc,treatment_sc,tool_sc,mask_sc,eye_sc]
sephora = pd.concat(frames)
sephora['price']=(sephora['highprice']+sephora['lowprice'])/2

sephora.to_csv('sephora.csv')



sephora












```

```python
avg=sephora.groupby(['subcat', 'cat']).agg('mean').drop(['lowprice','highprice'], axis = 1)
avg=avg
skincare=sephora.groupby('subcat').agg('mean').drop(['lowprice','highprice'], axis = 1).drop(['eye','lip','face','cheek'],axis=0)
skincare

makeup=sephora.groupby('subcat').agg('mean').drop(['lowprice','highprice'], axis = 1).drop(['cleanser','eye_sc','mask','moisturizer','tool','treatment'],axis=0)

avg.to_csv('avg.csv')

makeup
makeup.to_csv('makeup.csv')
skincare.to_csv('skincare.csv')
makeup=makeup
skincare=skincare
avg.to_csv('avg.csv')
```

```python
top10_price=sephora.groupby(['brandname','subcat']).agg('mean').sort_values(by='price', ascending=False).drop(['lowprice','highprice'],axis=1).head(10)
top10_rating=sephora.groupby(['brandname','subcat']).agg('mean').sort_values(by='rate', ascending=False).drop(['lowprice','highprice'],axis=1).head(10)
print(top10_rating)

top10_price.to_csv('top10_price.csv')
top10_rating.to_csv('top10_rating.csv')


bottom10_price=sephora.groupby(['brandname','subcat']).agg('mean').sort_values(by='price', ascending=True).drop(['lowprice','highprice'],axis=1).head(10)
bottom10_rating=sephora.groupby(['brandname','subcat']).agg('mean').sort_values(by='rate', ascending=True).drop(['lowprice','highprice'],axis=1).head(10)


bottom10_price.to_csv('bottom10price.csv')
bottom10_rating.to_csv('bottom10rating.csv')

print(top10_price)
print(top10_rating)
```
